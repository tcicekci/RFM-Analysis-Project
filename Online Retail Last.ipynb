```python
import numpy as np
import pandas as pd
import datetime as dt
```

```python
df = pd.read_excel(r"/Users/tugrulcicekci/Desktop/Portf√∂y Projeleri/2.RFM/OnlineRetail.xlsx")
df.head()
```

Data Cleaning by Removing Trailing Spaces and Null Handling

```python
df.columns = df.columns.str.strip()
df.dropna(subset=["InvoiceNo", "StockCode", "Description", "Quantity", "InvoiceDate", "UnitPrice", "CustomerID", "Country"], inplace=True)
df.shape
```

Adding New Columns for further Analysis

```python
snapshot_date = df['InvoiceDate'].max()
snapshot_date
```

```python
df["MostRecentPurchase"] = df.groupby('CustomerID')['InvoiceDate'].transform('max')
df["FirstPurchase"] = df.groupby("CustomerID")["InvoiceDate"].transform("min")
df['MostRecentPurchase'] = pd.to_datetime(df['MostRecentPurchase'])
df['FirstPurchase'] = pd.to_datetime(df['FirstPurchase'])
df["TimeSpan"] = df["MostRecentPurchase"] - df["FirstPurchase"]
df.head()
```

```python
df["TotalPrice"] = df["Quantity"] * df["UnitPrice"]
df["TotalSpending"] = df.groupby("CustomerID")["TotalPrice"].transform("sum")
df["TotalOrders"] = df.groupby("CustomerID").transform("size")
df[150:200]
```

```python
df['TimeSpanInDays'] = (df['MostRecentPurchase'] - df['FirstPurchase']).dt.days

# Calculate Frequency
df["TimeSpanInDays"] = df["TimeSpanInDays"] + 1
df['Frequency'] = (df['TotalOrders'] / df['TimeSpanInDays']).round(1)
df.head()
```

```python
#df.drop("TimeSpan", axis=1, inplace=True)
df.drop("InvoiceTime", axis=1, inplace=True)
df.head()
```

Calculating RFM Scores and Segmentation of Customers

```python
#Looking for values in the columns
df["Frequency"].unique()
```

```python
df["Frequency"].max()
df["Frequency"].min()
```

```python
df.head()
```

```python
df["RecencyInDays"] = (snapshot_date - df["MostRecentPurchase"]).dt.days
df.head()
```

```python
df.drop("Recency", axis=1, inplace=True)
df.head()
```

RFM Scores

```python
#Burada iptal edilmi≈ü sipari≈ülere dair hata olabilir
df['R_Score'] = pd.qcut(df['RecencyInDays'], 5, labels=[5, 4, 3, 2, 1]).astype(int)

# Score Frequency (higher = better)
df['F_Score'] = pd.qcut(df['Frequency'].rank(method='first'), 5, labels=[1, 2, 3, 4, 5]).astype(int)

# Score Monetary (higher = better)
df['M_Score'] = pd.qcut(df['TotalSpending'], 5, labels=[1, 2, 3, 4, 5]).astype(int)

df['RFM_Score'] = (
    df['R_Score'].astype(str) +
    df['F_Score'].astype(str) +
    df['M_Score'].astype(str)
)

df[150:200]
```

```python
def rfm_segment(row):
    r, f, m = row['R_Score'], row['F_Score'], row['M_Score']
    if r >= 4 and f >= 4 and m >= 4:
        return 'Champions'
    elif r >= 4 and f >= 3:
        return 'Loyal Customers'
    elif r >= 3 and m >= 4:
        return 'Big Spenders'
    elif r <= 2 and f <= 2 and m <= 2:
        return 'At Risk'
    else:
        return 'Others'

df['Segment'] = df.apply(rfm_segment, axis=1)
df["IsOrderCancelled"] = df["InvoiceNo"].astype(str).str.startswith("C")
df.head()
```

```python
df.drop("IsCancelled", axis=1, inplace=True)
```

```python
df.drop(["R_Score", "F_Score", "M_Score"], axis=1, inplace=True)
df.head()
```

```python
df['CustomerID'] = df['CustomerID'].astype('Int64')
df.head()
```

```python
df.to_csv("online_retail_last", index=False)
```
